```python
#Removing null values
train_data = train_data.dropna()
```

```python
#Defining X and Y
X = train_data.drop('price_range',axis=1)
Y = train_data['price_range']
```

```python
#Splitting data into training and cros_validation data
from sklearn.model_selection import train_test_split

# First split: 70% training, 30% temporary (will be split further)
X_train, X_temp, Y_train, Y_temp = train_test_split(X, Y, test_size=0.30, random_state=42)

# Second split: 50% of the temporary set for validation, 50% for testing
X_csv, X_test, Y_csv, Y_test = train_test_split(X_temp, Y_temp, test_size=0.50, random_state=42)

# Now we have:
# X_train, Y_train: 70% of the original data for training
# X_csv, Y_csv: 15% of the original data for validation
# X_test, Y_test: 15% of the original data for testing


print(X_train.shape)
print(X_csv.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_csv.shape)
print(Y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_csv = sc.transform(X_csv)
X_test = sc.transform(X_test)
```

saving the scaler

```python
import joblib

X_train_scaled = X_train

# Save the fitted scaler
joblib.dump(sc, 'standard_scaler.pkl')
```

supervised machine learning: logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,Y_train)
```

test the cross validation data set

```python
pv1 = lr.predict(X_csv)
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
a1 = accuracy_score(Y_csv,pv1)
print('Accuracy Score:',a1)
print('\nConfusion Matrix:\n',confusion_matrix(Y_csv,pv1))
print('\nClassification Report:\n',classification_report(Y_csv,pv1))
```

Hyperparameter Tuning with GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'C': [0.1, 1, 10, 100],  # Inverse of regularization strength
    'penalty': ['l1', 'l2'],  # Regularization type
    'solver': ['liblinear', 'saga']  # Solver for optimization
}

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=lr, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# Fit the GridSearchCV object on the training data
grid_search.fit(X_train, Y_train)

# Get the best parameters and the best model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Evaluate the best model on the cross-validation data
Y_pred_best = best_model.predict(X_csv)
accuracy_best = accuracy_score(Y_csv, Y_pred_best)
print('Best Parameters:', best_params)
print('Best Model Accuracy on Cross-Validation Data:', accuracy_best)
print('Confusion Matrix:\n', confusion_matrix(Y_csv, Y_pred_best))
print('Classification Report:\n', classification_report(Y_csv, Y_pred_best))
```

test the test data

```python
# Evaluate the best model on the test data
Y_pred_test = best_model.predict(X_test)
accuracy_test = accuracy_score(Y_test, Y_pred_test)
print('Final Model Accuracy on Test Data:', accuracy_test)
print('Confusion Matrix on Test Data:\n', confusion_matrix(Y_test, Y_pred_test))
print('Classification Report on Test Data:\n', classification_report(Y_test, Y_pred_test))
```

Optimize the model:

â–  Choosing an appropriate algorithm to make the result good enough

| Algorithm        | Speed (Training Time) | Expected Accuracy | Resource Extensive (Memory & CPU) | Best Use Case |
|------------------|-----------------------|-------------------|-----------------------------------|---------------|
| Logistic Regression | Fast                  | High              | Low                               | Binary classification, linear problems |
| SVM              | Moderate              | High              | High                              | Non-linear classification, small to medium datasets |
| Random Forest    | Moderate              | Very High         | High                              | Ensemble methods, feature importance, non-linear problems |
| KNN Classifier   | Fast                  | Moderate          | Moderate                          | Small datasets, non-linear problems, instance-based learning |

As you see in the table above, the SVM and Random Forest algorithms are high resource extensive with moderate speed.

The best use case for them is non-linear and more complicated data and features, so the best use case for them is to capture complex related data.

With comparing all of the above algorithms, I find the logistic regression algorithm is more suitable in every aspect, and it's the best use case for a project like this with Final Model Accuracy of 97%.

I used scikit-learn instead of more complex neural networks with TensorFlow because it's the fastest and simplest, and it's suitable in this situation.

```python
# Load the test data
test_data = pd.read_csv("test.csv")

# For example, if you need to scale the features
X_test = test_data.drop('id', axis=1)  # Drop 'id'
X_test = sc.transform(X_test)  # Use the same scaler used during training
ids = test_data['id']

# Select the first 10 examples
X_test_10 = X_test[:10]
ids_10 = ids[:10]

# Predict the price range for the first 10 examples
predictions = best_model.predict(X_test_10)

# Get the probability estimates for each class
probabilities = best_model.predict_proba(X_test_10)

# Format the output
for i, (pred, prob, mobile_id) in enumerate(zip(predictions, probabilities, ids_10)):
    confidence = max(prob)  # Get the highest probability (confidence) for the predicted class
    print(f"mobile_{mobile_id} : predicted cost is {pred} with confidence {confidence:.2f}")
```
